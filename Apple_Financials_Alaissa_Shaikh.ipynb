#### **Cleaning dataset columns accordingly**





```python
# Defining a function to clean columns with financial data
def clean_column(column):
    return pd.to_numeric(apple_data[column].replace('[\$,]', '', regex=True).replace(',', '', regex=True))
```

```python
# Cleaning the "Gross Margin" column separately to remove percentage signs
apple_data["Gross Margin"] = apple_data["Gross Margin"].str.replace('%', '', regex=True).astype(float)
```

```python
# Cleaning other numeric columns with dollar signs or commas
columns_to_clean = [
    "EBITDA (millions)", "Revenue (millions)", "Gross Profit (millions)",
    "Op Income (millions)", "Net Income (millions)", "EPS",
    "Shares Outstanding", "Total Assets (millions)", "Cash on Hand (millions)",
    "Long Term Debt (millions)", "Total Liabilities (millions)"
]

for column in columns_to_clean:
    apple_data[column] = clean_column(column)
```

#### **Stock Price vs EPS and Net Income**

EPS (USD) is usually much smaller, often in single digits.
If both were plotted on the same y-axis, one variable would dominate the graph, making it hard to interpret the other.

2.  Left y-axis (blue): Represents the stock price, making it easier to observe trends in Apple's market value.

3.   Right y-axis (green): Represents the EPS, showing how Apple’s earnings relate to its stock price.

This approach helps to visualize how stock price movements correlate with the company’s profitability (EPS) over time, even though their numerical ranges are quite different. It provides a clearer and more insightful comparison between these two metrics.

#### **Trend Decomposition**

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decomposing Revenue Trend
result = seasonal_decompose(apple_data["Revenue (millions)"], model='multiplicative', period=1)

# Plotting the decomposition
result.plot()
plt.show()
```

Seasonal Decomposition: The seasonal_decompose function decomposes a time series into three components:

1.   Trend: The underlying trend or long-term progression in the data.
2.   Seasonal: The repeating short-term fluctuations or patterns within the data.
3.   Residual: The noise or leftover component after removing trend and seasonality.

In this case, the decomposition is applied to Revenue (millions) data using a multiplicative model (where the components are multiplied together).

This model is often used when seasonal fluctuations increase in magnitude over time. The period=1 indicates yearly data (i.e., each year is a data point).

Plotting: The result.plot() function generates a plot showing:

1.   The original data (Revenue),
2.   The trend component (long-term direction of the revenue),
3.   The seasonal component (yearly patterns),
4.   The residual component (remaining fluctuations).


Purpose:
The decomposition allows you to better understand the underlying patterns in Apple’s revenue data, such as long-term growth trends, seasonality, and any irregular fluctuations.

By visualizing the separate components, you can identify if revenue growth is primarily driven by long-term trends, periodic patterns, or random factors.

#### **Revenue Forecasting**

```python
from statsmodels.tsa.arima.model import ARIMA

# Fitting an ARIMA model for Revenue
revenue = apple_data["Revenue (millions)"]
model = ARIMA(revenue, order=(1, 1, 1))  # Simple ARIMA parameters
model_fit = model.fit()
```

```python
# Forecasting future revenue
forecast = model_fit.forecast(steps=5)
future_years = range(2025, 2030)
```

```python
# Plotting Revenue and Forecast
plt.figure(figsize=(10, 6))
plt.plot(apple_data["year"], revenue, label="Historical Revenue", marker='o')
plt.plot(future_years, forecast, label="Forecasted Revenue", marker='o', linestyle='--')

plt.title("Apple Revenue Forecast (2025-2030)")
plt.xlabel("Year")
plt.ylabel("Revenue (millions/USD)")
plt.legend()
plt.grid(True)
plt.show()
```

Explanation:

ARIMA Model: The ARIMA model is applied to the Revenue data to capture both the autoregressive (AR) and moving average (MA) components, along with any trends in the data:

The order=(1, 1, 1) parameters indicate the model uses:
1.   1 autoregressive term (AR): Relates current values to previous values.
2.   1 difference term (I): Makes the series stationary by subtracting previous values.
3.   1 moving average term (MA): Accounts for the relationship between current values and past forecast errors.


Forecasting: After fitting the model on the historical revenue data, the forecast function predicts the revenue for the next 5 years (2025-2030).

Plotting:

1.   The blue line represents the historical revenue (2009-2024).
2.   The dashed red line represents the forecasted revenue (2025-2030), showing the model's estimate of future revenue trends.


Purpose:

1.   Forecasting: The ARIMA model is used to predict future revenue, which can help make informed decisions about Apple's financial outlook.
2.   Trend Analysis: The plot provides a visual comparison between historical performance and future projections, helping to assess how Apple's revenue might evolve.


This type of analysis is useful for understanding potential future growth trends and making strategic business decisions based on forecasted data.

#### **Rooling averages for smoother trends**

```python
# Calculating Rolling Averages
apple_data["Revenue Rolling Avg"] = apple_data["Revenue (millions)"].rolling(window=3).mean()
apple_data["Net Income Rolling Avg"] = apple_data["Net Income (millions)"].rolling(window=3).mean()

# Plotting Rolling Averages
plt.figure(figsize=(10, 6))
plt.plot(apple_data["year"], apple_data["Revenue Rolling Avg"], label="Revenue Rolling Avg (3 years)", marker='o')
plt.plot(apple_data["year"], apple_data["Net Income Rolling Avg"], label="Net Income Rolling Avg (3 years)", marker='o')

plt.title("Apple Revenue and Net Income Rolling Averages (2009-2024)")
plt.xlabel("Year")
plt.ylabel("Value (millions/USD)")
plt.legend()
plt.grid(True)
plt.show()
```

Rolling Averages:

A rolling average smooths out short-term fluctuations by averaging data points over a specific window of time. In this case, a 3-year rolling average is applied, which averages the values of each year with the previous two years.

This helps to reduce the impact of year-to-year variability and gives a clearer picture of longer-term trends in revenue and net income.

Plotting:

1. The blue line represents the 3-year rolling average for Revenue.
2. The orange line represents the 3-year rolling average for Net Income.
By plotting both rolling averages, we can compare the trends in both revenue and net income over time.


Purpose:
1.   Smoothing Trends: The rolling averages make it easier to identify the underlying trends in Apple's revenue and net income by filtering out noise or short-term fluctuations.
2.   Long-Term Insights: This approach helps to assess the company’s overall financial health and growth trajectory over multiple years, making it easier to detect patterns such as consistent growth or potential slowdowns.


This type of analysis is useful for evaluating the stability and consistency of Apple's financial performance over time.
